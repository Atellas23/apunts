```python
import matplotlib.pyplot as plt

# read the file with the wordcount and plot it
# the file has format
# <abs_freq>, <word_1>
# <abs_freq>, <word_2>
# <abs_freq>, <word_3>
# ...................
# <abs_freq>, <word_n>

word_freq = []
with open('input.txt', newline='', encoding='utf-16') as file:
    for line in file:
        contents = line.split(',')
        word_freq.append(int(contents[0]))

word_freq.reverse()
fig, ax = plt.subplots(1, 2, figsize=(18, 7))
ax[0].plot(word_freq)
ax[0].set_xlabel('Rank')
ax[0].set_ylabel('Absolute Frequency')
ax[0].set_title('Rank vs Absolute Frequency')

ax[1].loglog(word_freq)
ax[1].set_xlabel('Rank')
ax[1].set_ylabel('Absolute Frequency')
ax[1].set_title('Rank vs Absolute Frequency (log-log plot)')

fig.tight_layout()
fig.show()
```
